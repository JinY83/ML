(2) 데이터 분리 : 학습 & 테스트 데이터 변수 지정

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, 
                                                    test_size=0.3, random_state=10)

# 관용적으로 X 는 대문자, y 는 소문자로 작성한다

####################3 pdf 83p
# test sixe : 전체 데이터에서 테스트 데이터 세트 크기를 얼마로 샘플링 할것인가를 결정
# train size :   
# random_state 설명 : https://www.inflearn.com/questions/110501
```

```python
from sklearn.tree import DecisionTreeClassifier

### DecisionTreeClassifier
# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html?highlight=decisiontreeclassifier#sklearn.tree.DecisionTreeClassifier

# 파라메터 ㅣ(*, criterion='gini', splitter='best', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=None, random_state=None, max_leaf_nodes=None, min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0)
## => max_depth : The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.
```

```python
from sklearn.metrics import accuracy_score

### accuracy_score
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html?highlight=accuracy_score#sklearn.metrics.accuracy_score

### 파라메터 : (y_true, y_pred, *, normalize=True, sample_weight=None)
# y_true : Ground truth (correct) labels
# y_pred : Predicted labels, as returned by a classifier.
# normalize : If False, return the number of correctly classified samples. Otherwise, return the fraction of correctly classified samples.
# sample_weight : Sample weights.
```

```python
## 예측 정확도 출력

df_clf = DecisionTreeClassifier(max_depth=4, random_state=11)
df_clf.fit(X_train, y_train)
pred = df_clf.predict(X_test)

print(f'예측정확도:{accuracy_score(pred, y_test)}')
```

```python
iris = load_iris()
df_clf = DecisionTreeClassifier()
train_data = iris.data
train_label = iris.target
df_clf.fit(train_data, train_label)

## 학습 데이터 셋으로 예측 수행
pred = df_clf.predict(train_data)
print('예측 정확도:', accuracy_score(pred, train_label))
```

```python
## 매개변수값의 변화에 따라 예측 정확도가 어떻게 변화하는지 확인

df_clf = DecisionTreeClassifier(max_depth=1, random_state=11)     # max_depth 조정 : 예측정확도 떨어짐
df_clf.fit(X_train, y_train)
pred = df_clf.predict(X_test)

print(f'예측정확도:{accuracy_score(pred, y_test)}')
```

*** parameter
 1. parameter (매개변수, 파라미터) : 신경망의 가중치처럼 학습 과정에서 얻어지는 값
 2. hyper-parameter (초매개변수) : 학습률이나 배치 크기처럼 사용자가 임의로 지정해야 하는 값

학습 데이터는 모델의 파라미터를 결정하는 데 사용하고, 검증 데이터는 학습 횟수(Number of Epochs), 학습률(Learning Rate) 등 하이퍼파라미터라고 불리는 학습 설정값들을 다양하게 조합하여 모델들을 학습시킨 후에 성능 평가를 통해 가장 성능이 높은 모델을 선택하는 데 사용하며, 테스트 데이터는 최종적으로 결정된 모델의 성능을 수치화하는 데 사용합니다.\\

지금 구현한 프레임워크를 가지고 다양한 ‘초매개변수’(계층 크기, 학습 속도, 배치 크기 등)를 활용해 테스트해 보자. 적절한 지형(다른 ‘초매개변수’와 함께)을 얻으려면 수많은 변경과 테스트 과정이 필요 하다. 입력과 출력 계층의 크기는 용도에 따라 달라질 수 있지만(예를 들어, 분류의 경우 입력 크기는 이미지의 픽셀 개수이고 출력 크기는 예측할 클래스 개수가 될 것이다), 은닉 계층은 신중하게 설계돼야 한다.

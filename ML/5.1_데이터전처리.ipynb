(1) 데이터 인코딩

sklearn.preprocessing.LabelEncoder

a. import 방법 : from sklearn import preprocessing

b. 공식홈 : https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html?highlight=labelen#sklearn.preprocessing.LabelEncoder


```python
from sklearn.preprocessing import LabelEncoder
```

```python
items = ['TV', '냉장고', '전자렌지', '컴퓨터', '선풍기', '선풍기', '믹서', '믹서']

# Label Encoder 를 객체로 생성한 후 fit() 과 transform() 으로 label 인코딩 수행
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
print('인코딩 변환값:', labels)
print('인코딩 클래스: ', encoder.classes_)
print('디코딩 원본 값: ', encoder.inverse_transform([4,5,2,0,1,1,3,3]))
```

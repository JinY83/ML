(1) k-fold


* k fold 설명 \

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
iris = load_iris()
features = iris.data
label = iris.target

X_train, X_test, y_train, y_test = train_test_split(iris.data, 
                                                    iris.target, 
                                                    test_size=0.3, 
                                                    random_state=10)
```

```python
kfold = KFold(n_splits=5)
cv_accuracy = []
print('붓꽃 데이터 세트 크기 :', features.shape[0])
```

```python
df_clf = DecisionTreeClassifier(max_depth=1, random_state=11)     # max_depth 조정 : 예측정확도 떨어짐
df_clf.fit(X_train, y_train)
pred = df_clf.predict(X_test)
```

```python
## k fold

n_iter = 0
# KFold 객체의 split() 호출하면 폴드 별 학습용 검증용 테스트의 로우 인덱스를 array로 변환
for train_index, test_index in kfold.split(features):
    # KFold split() 으로 반환된 인덱스를 이용하여 학습용, 검증용 데이터 추출
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    # 학습 및 예측
    df_clf.fit(X_train, y_train)
    pred = df_clf.predict(X_test)
    n_iter += 1
    # 반복시마다 정확도 측정
    accuracy = np.round(accuracy_score(y_test,pred), 4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print('\n#{0} 교차 검증 정확도: {1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'.format(n_iter, accuracy, train_size, test_size))
    print('#{0} 검증 세트 인덱스: {1}'.format(n_iter, test_index))
    cv_accuracy.append(accuracy)
print('\n## 평균 검증 정확도:', np.mean(cv_accuracy))
```

(2) Stratified k-fold

* k-fold 와 stratified fold

```python
import pandas as pd

iris_data = load_iris()
iris_data.keys()
df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
df['label'] = iris_data.target
```

```python
df['label'].value_counts()
```

```python
## 아래 for 문은 k fold 방식임
 
kfold = KFold(n_splits=3)
dt_clf = DecisionTreeClassifier(random_state=121)

n_iter = 0
for train_index , test_index in kfold.split(features):
    n_iter +=1
    label_train = df['label'].iloc[train_index]
    label_test = df['label'].iloc[test_index]
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    print(f'{n_iter} : {label_train.value_counts()} \n')
    print(f'{n_iter} : {label_test.value_counts()}')
    dt_clf.fit(X_train,y_train )
    pred = dt_clf.predict(X_test)
    accuracy = np.round(accuracy_score(y_test,pred), 4)
    
    print(pred)
    print(accuracy)
    
```

```python
from sklearn.model_selection import StratifiedKFold
import numpy as np

sk_fold = StratifiedKFold(n_splits=3)
n_iter +=0
val_list = []
for train_index , test_index in sk_fold.split(features, label):
    n_iter +=1
    label_train = df['label'].iloc[train_index]
    label_test = df['label'].iloc[test_index]
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    print(f'{n_iter} : {label_train.value_counts()} \n')
    print(f'{n_iter} : {label_test.value_counts()}')
    dt_clf.fit(X_train,y_train )
    pred = dt_clf.predict(X_test)
    print(pred)
    accuracy = np.round(accuracy_score(y_test, pred), 4)
    print(accuracy)
    val_list.append(accuracy)
val_acc = np.mean(val_list)
```

---------------------------------------------------
** parameters & 작성법\
 : (estimator, X, y=None, *, groups=None, scoring=None, cv=None, n_jobs=None, verbose=0, fit_params=None, pre_dispatch='2*n_jobs', error_score=nan

** parameter 설명 (1,2,5 는 반드시 작성해 주어야 함)
1. estimator : 분류(classifier)인지 회귀(regressor) 인지 구분하는 인자 - 반드시 작성해 주어야 함
2. x : feature 데이터 세트
3. y : label 데이터 세트 
4. scoring : 예측 성능 평가지표 = accuracy, ... 등 지정해줌
 -  단일 값만 반환해야 하는 서명 scorer(estimator, X, y)가 있는 str(모델 평가 문서 참조) 또는 스코어러 호출 가능 객체/함수. cross_validate와 유사하지만 단일 메트릭만 허용
5. cv : 교차검증 폴드 수 (생략할 경우 default = 5)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
dt_clf = DecisionTreeClassifier(random_state=156)

data = iris_data.data
label = iris_data.target

# 성능 지표는 '정확도(accuracy)', 교차 검증 세트(cv)는 '3개'
scores = cross_val_score(dt_clf, data, label, scoring= 'accuracy', cv=3)
print('교차 검증별 정확도 :', np.round(scores, 4))
print('평균 검증 정확도 :', np.round(np.mean(scores), 4))
```

(4) Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
# 데이터 로딩하고 학습 / 테스트 데이터 분리
iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, 
                                                    iris_data.target, 
                                                    test_size=0.2, 
                                                    random_state=121)

dtree = DecisionTreeClassifier()

# paramater 들을 딕셔너리 형태로 설정
parameters = {'max_depth':[1,2,3], 'min_samples_split':[2,3]}
```

```python
import pandas as pd

# param_grid 의 하이퍼파라미터들을 3개의 train, test set fold 로 나누어서 테스트 수행 설정
## refit : default 는 True -> true 일 때 가장 좋은 파라미터 설정으로 재학습 시킴
grid_dtree = GridSearchCV(dtree, param_grid= parameters, cv= 3, refit= True)

# 붓꽃 train 데이터로 param_grid 의 하이퍼 파라미터들을 순차적으로 학습/평가
grid_dtree.fit(X_train, y_train)

# 출력 형태 확인
print(grid_dtree.cv_results_)      

# Grid~ cv 결과를 추출해서 데이터프레임으로 변환 
# - 정형화된 형태로 안 봐도 되면 이 과정은 생략해도 된다
scores_df = pd.DataFrame(grid_dtree.cv_results_)
scores_df[['params', 'mean_test_score', 'rank_test_score', 
            'split0_test_score', 'split1_test_score', 'split1_test_score']]
```

```python
print(grid_dtree.best_params_)    # 최적의 파라미터 확인
print(grid_dtree.best_score_)     # 제일 높은 정확도 확인
```

```python
# Grid~ cv refit 로 이미 학습된 estimator 반환
estimator = grid_dtree.best_estimator_

pred = estimator.predict(X_test)
print(accuracy_score(y_test, pred))      # 테스트데이터 세트의 정확도 확인
```
